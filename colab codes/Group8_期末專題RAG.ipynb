```python
from pymongo import MongoClient
import pandas as pd
```

```python
url = userdata.get('MGDB_IAI')
client = MongoClient(url)
db = client["Group8"]
hanzi_collection = db["漢字與詞語"]
training_collection = db["labeled_training_data"]
```

```python
hanzi_path="/content/gdrive/MyDrive/114-2 AI導論/期末專題/級數資料/漢字與詞語.csv"
text_path="/content/gdrive/MyDrive/114-2 AI導論/期末專題/訓練資料文本/labeled_training_data_文本.csv"
poem_path="/content/gdrive/MyDrive/114-2 AI導論/期末專題/訓練資料文本/labeled_training_data_詩集.csv"
```

```python
df_hanzi=pd.read_csv(hanzi_path)
df_text=pd.read_csv(text_path)
df_poem=pd.read_csv(poem_path)
```

```python
# add source label
df_text["source"]="文本"
df_poem["source"]="詩集"
```

```python
# convert to Mongo format
hanzi_records=df_hanzi.to_dict("records")
text_records=df_text.to_dict("records")
poem_records=df_poem.to_dict("records")
```

```python
# upload to MongoDB
hanzi_collection.insert_many(
    hanzi_records
)
training_collection.insert_many(
    text_records
)
training_collection.insert_many(
    poem_records
)
```

```python
# check result
print("漢字與詞語：",hanzi_collection.count_documents({}))
print("training_data：",
      training_collection.count_documents({}))
```
